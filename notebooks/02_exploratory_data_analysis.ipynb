This notebook uses the `EDAOrchestrator` to perform a comprehensive exploratory analysis of the Home Credit Default Risk dataset.

```python
import sys
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

# Add src to path
PROJECT_ROOT = Path.cwd().parent
sys.path.append(str(PROJECT_ROOT / "src"))

from data.loader import DatasetLoader
from eda import EDAOrchestrator
```

```python
# Setup paths
data_dir = PROJECT_ROOT / "data" / "raw"
reports_dir = PROJECT_ROOT / "reports"

print("Loading data...")
loader = DatasetLoader(data_dir)
df = loader.load_csv("application_train.csv")
print(f"Dataset shape: {df.shape}")
```

```python
# Sampling for interactive speed (optional)
df_sample = df.sample(min(50000, len(df)), random_state=42)
print(f"Sample size for EDA: {len(df_sample)}")
```

```python
# Initialize Orchestrator
orchestrator = EDAOrchestrator(output_dir=reports_dir, target_column="TARGET")

# Define features to analyze (or let it auto-detect)
num_cols = [
    "AMT_INCOME_TOTAL", "AMT_CREDIT", "AMT_ANNUITY", "AMT_GOODS_PRICE", 
    "DAYS_BIRTH", "DAYS_EMPLOYED", "DAYS_REGISTRATION", "DAYS_ID_PUBLISH",
    "EXT_SOURCE_1", "EXT_SOURCE_2", "EXT_SOURCE_3"
]
cat_cols = [
    "NAME_CONTRACT_TYPE", "CODE_GENDER", "FLAG_OWN_CAR", "FLAG_OWN_REALTY", 
    "NAME_TYPE_SUITE", "NAME_INCOME_TYPE", "NAME_EDUCATION_TYPE", 
    "NAME_FAMILY_STATUS", "NAME_HOUSING_TYPE", "OCCUPATION_TYPE"
]

print("Running EDA experiments...")
orchestrator.run(df_sample, num_features=num_cols, cat_features=cat_cols)
```

```python
# View high-level target relationships
target_rel = pd.read_csv(reports_dir / "target_relationships.csv")
display(target_rel.sort_values("mutual_info", ascending=False).head(10))
```

```python
# View feature insights
insights = pd.read_csv(reports_dir / "feature_insights.csv")
display(insights.head(10))
```
